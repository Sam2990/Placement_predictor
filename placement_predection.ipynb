```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
# from sklearn.externals import j
import joblib
import tkinter
import tkinter.font as font
```

```python
df=pd.read_csv('Placement_Data_Full_Class.csv')
# df.head()
df.tail()
```

```python
df=df.drop(columns=['sl_no','salary'])
df.head()
```

```python
df["ssc_b"].unique()
```

```python
df['ssc_b'] = df["ssc_b"].map({'Central':1,"Others":0})
df.head()
```

```python
df["hsc_b"].unique()
df['hsc_b']=df["hsc_b"].map({"Others":0,"Central":1})
df.head()
```

```python
df["hsc_s"].unique()
df['hsc_s'] = df['hsc_s'].map({"Commerce":1,"Science":2,"Arts":3})
df.head()
```

```python
df['gender'].unique()
df['gender']=df['gender'].map({'M':1,'F':2})
df.head()
```

```python
df["degree_t"].unique()
df['degree_t']=df['degree_t'].map({'Sci&Tech':1, 'Comm&Mgmt':2, 'Others':3})
df.head()
```

```python
df['workex'].unique()
df['workex']=df['workex'].map({'No':0, 'Yes':1})
```

```python
df['specialisation'].unique()
df['specialisation']=df['specialisation'].map({'Mkt&HR':1, 'Mkt&Fin':2})
```

```python
df['status'].unique()
df['status']=df['status'].map({'Placed':1, 'Not Placed':0})
df.head()
# df.isnull().sum()
```

```python
x=df.drop(columns="status")
y=df['status']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)
```

```python
lr = LogisticRegression()
model=lr.fit(x_train,y_train)
y_pridict=model.predict(x_test)
accuracy_score(y_pred=y_pridict,y_true=y_test)
```

```python
joblib.dump(lr,'placement_predection_oo')
```
